Generate normal random variables from the uniform distribution using the Box-Muller method. This is only a fun exercise as the numpy functions are optimised for generating normal random variables.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
# set to plot automatically
%matplotlib inline
```

First create two independent random variables U~U(0,1) and V~U(0,1) 

```python
random_uniform_u = np.random.uniform(0,1,1000)
random_uniform_v = np.random.uniform(0,1,1000)
```

Then use the Box-Muller method to create two sets of independent standard normal random variables. The Box-Muller method says that you can create independent standard normal variables via the following formulas: $X_1 := \sqrt{(-2\log{U})}\cos{(2\pi V)}$	and $X_2 := \sqrt{(-2\log{U})}\sin{(2\pi V)}$

```python
random_normal_1 = np.multiply(np.sqrt(-2*np.log(random_uniform_u)),np.cos(2*pi*random_uniform_v))
```

```python
random_normal_2 = np.multiply(np.sqrt(-2*np.log(random_uniform_u)),np.sin(2*pi*random_uniform_v))
```

Plot the results

```python
sns.distplot(random_normal_1);
sns.distplot(random_normal_2);
```

```python
fig = plt.figure(figsize=(6,6));
plt.scatter(random_normal_1,random_normal_2);
plt.xlabel('$X_1$');
plt.ylabel('$X_2$');
```
