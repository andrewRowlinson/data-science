```python
from mplsoccer import VerticalPitch, Radar
p = VerticalPitch(pitch_type='uefa')
radar = Radar(params=['Agility', 'Speed', 'Strength'], min_range=[0, 0, 0], max_range=[10, 10, 10])
fig, ax = p.draw(figsize=(6, 9))
axs = p.formation('433', kind='axes', width=15, aspect=1, ax=ax)
for player_ax in axs:
    radar.setup_axis(ax=axs[player_ax])
    rings_inner = radar.draw_circles(ax=axs[player_ax], facecolor='#ffb2b2', edgecolor='#fc5f5f')
    values = [5, 3, 10]
    radar_poly, vertices = radar.draw_radar_solid(values, ax=axs[player_ax], kwargs={'facecolor': '#00f2c1', 'alpha': 0.6})
    axs[player_ax].set_facecolor('None')
```
