```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
p = Pitch()
```

```python
figwidth, figheight = 16, 9
fig = plt.figure(figsize=(figwidth, figheight))
blank_axes = fig.add_axes((0, 0, 1, 1))
#blank_axes.axis('off')
pitch_height = 0.2
pitch_width = pitch_height * p.ax_aspect * figheight / figwidth
ax = fig.add_axes((0.1, 0.1, pitch_width, pitch_height))
p.draw(ax=ax)
ax2 = fig.add_axes((0.6, 0.6, pitch_width, pitch_height))
p.draw(ax=ax2)
```
