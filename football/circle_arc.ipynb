```python
from mplsoccer import Pitch, Standardizer
import numpy as np
```

```python
pitch = Pitch(pitch_type='statsbomb')
```

```python
standard = Standardizer(pitch_to=pitch.pitch_type, pitch_from='uefa')
```

```python
x_25, _ = standard.transform([105-25],[68/2])
```

```python
x_25_statsbomb = pitch.dim.right - x_25
```

```python
x_25_statsbomb
```

```python
x = np.random.uniform(0, 120, 5000)
y = np.random.uniform(0, 80, 5000)
```

```python
angle, distance = pitch.calculate_angle_and_distance(x, y, [pitch.dim.right] * len(x),
 [pitch.dim.center_width] * len(y))
```

```python
mask_25 = distance < x_25_statsbomb
fig, ax = pitch.draw()
pitch.scatter(x[mask_25], y[mask_25], ax=ax, alpha=0.3)
```
