```python
from mplsoccer import Sbopen, Pitch
```

```python
parser = Sbopen()
```

```python
df_match = parser.match(1267, 107)
```

```python
df_events, df_related, df_freeze, df_tactic  = parser.event(3923881)
```

```python
mask_shot = df_events.type_name == 'Shot'
mask_nigeria = df_events.team_name == 'Nigeria'
mask_cote = df_events.team_name == "Côte d'Ivoire"
```

```python
df_events.groupby('team_name').shot_statsbomb_xg.sum()
```

```python
p = Pitch()
fig, ax = p.draw()
p.scatter(df_events[mask_shot & mask_nigeria].x,
          df_events[mask_shot & mask_nigeria].y,
          s=df_events[mask_shot & mask_nigeria].shot_statsbomb_xg * 1000,
 ax=ax, color='#1B7439')
p.scatter(120 -df_events[mask_shot & mask_cote].x,
          80 - df_events[mask_shot & mask_cote].y,
          s=df_events[mask_shot & mask_cote].shot_statsbomb_xg * 1000,
 ax=ax, color='#EA7600')
```
