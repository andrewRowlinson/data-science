```python
import matplotlib.pyplot as plt
from mplsoccer import lines
```

```python
fig, ax = plt.subplots()
lines(0, 0, 0.5, 0.5, color='red', comet=True, transparent=True, alpha_start=0.4, alpha_end=1, n_segments=100, label='yo', ax=ax)
lines(0.2, 0.8, 0.6, 0, color='cornflowerblue', comet=True, transparent=True, alpha_start=0, alpha_end=1, n_segments=100, label='yoyo', ax=ax)
lines(0.5, 0.5, 1, 0.5, color='orange', comet=False, transparent=True, alpha_start=0, alpha_end=1, n_segments=100, label='yoyoyo', ax=ax)
lines(0, 1, 1, 0, cmap='viridis', comet=True, transparent=True, alpha_start=0, alpha_end=1, n_segments=100, label='yoyoyoyo', ax=ax)

ax.legend()
```
